#### briny & salty 간 상관관계

```python
import matplotlib.pyplot as plt

# Plot a scatter plot of briny vs salty
meta_data.plot.scatter(x="briny", y="salty")
meta_data.plot.scatter(x="rich", y="full_bodied")
# Show the plot
plt.show()
```

#### 결론
---
- briny & salty's correlation : 0.774629
- rich & full_bodied's correlation : 0.834311
- rich 와 bodied는 아주 강한 상관관계를 보이지만 다른 의미라는 의견으로 그 사이 무언가 상호작용이 있을 것으로 예상.
- briny와 salty는 아주 강한 상관관계를 보이며 의미상 유사하다. 
- briny 는 outlier가 많고 3사분위수까지 0인 데이터이기 때문에 제거를 수행합니다.
