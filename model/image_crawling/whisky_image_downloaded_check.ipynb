```python
import os
import csv
```

```python
file_list = os.listdir('whisky_tmp')
file_list
```

```python
store = set()
for file_name in file_list:
    no = file_name[3:file_name.find('.')]
    store.add(no)
    # whisky 번호 획득
print(store)
```

```python

# whisky 번호로 이름 획득
f_in = open('whisky.csv', 'r')
rdr = csv.reader(f_in)

f_out = open('output.csv', 'w', newline='')
wr = csv.writer(f_out)

for line in rdr:
    if line[0] == '' or line[0] in store:
        wr.writerow(line)
        print(line)
f_in.close()
f_out.close()
    
```
