```python
from rembg import remove
import os
```

```python

input_path = 'whisky_images/NO_1.png'
output_path = 'whisky_images_nobg/NO_1.png'

with open(input_path, 'rb') as i:
    with open(output_path, 'wb') as o:
        input = i.read()
        output = remove(input)
        o.write(output)
```

```python
file_list = os.listdir('whisky_images')
file_list
```

```python
for i, file_name in enumerate(file_list):
    input_path = 'whisky_images/' + file_name
    output_path = 'whisky_images_nobg/' + file_name
    
    with open(input_path, 'rb') as i:
        with open(output_path, 'wb') as o:
            input = i.read()
            output = remove(input)
            o.write(output)
            
            print(file_name)
```
