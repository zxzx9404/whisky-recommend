# 유사한 위스키 추천을 위해 위스키간 Cosine Similarity 를 측정합니다.
---
- 유사한 위스키는 다음과 같은 기준으로 정의합니다.
- Categorical Variables : category, price_tier
- Numeric Variables : abv, flavor(smoky ~ floral)

```python
import os
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

## item_features 준비

```python
item_features = pd.read_csv("../v4/dataset/item_features.csv", index_col=0, encoding="UTF-8")
```

```python
whisky = pd.read_csv("../dataset/whisky.csv", index_col = 0, encoding="UTF-8")
```

```python
item_features
```

## Get Cosine Similarity

```python
def get_similar_whisky(idx, item_features, k=5):
    ## get Cosine Similarity
    cosine_sim = cosine_similarity(item_features, item_features)
    
    ## get similarity scores
    scores = cosine_sim[idx]
    
    ## sort by similarity
    return np.argsort(-scores)[1:k+1]
```

```python
result = get_similar_whisky(94, item_features)
```
